```python
class GeneralNeuron:
   
    def __init__(self, firing_rate):
        self.firing_rate = firing_rate

    def activate(self, stimulus_strength):
        pass

class SensoryNeuron(GeneralNeuron):
    def __init__(self, firing_rate, receptor_type):
        super().__init__(firing_rate)
        self.receptor_type = receptor_type

    def sense_stimulus(self, stimulus):
        pass

class MotorNeuron(GeneralNeuron):
    def __init__(self, firing_rate, target_muscle):
        super().__init__(firing_rate)
        self.target_muscle = target_muscle

    def control_muscle(self):
        pass

class Photoreceptor(SensoryNeuron):
    def __init__(self, firing_rate):
        super().__init__(firing_rate, receptor_type="light")

    def light_detection(self, light_intensity):
        pass

class Mechanoreceptor(SensoryNeuron):
    def __init__(self, firing_rate):
        super().__init__(firing_rate, receptor_type="pressure")

    def pressure_detection(self, pressure_level):
        pass

class AlphaMotorNeuron(MotorNeuron):
    def __init__(self, firing_rate):
        super().__init__(firing_rate, target_muscle="skeletal muscle")

    def skeletal_muscle_control(self):
        pass

class GammaMotorNeuron(MotorNeuron):
    def __init__(self, firing_rate):
        super().__init__(firing_rate, target_muscle="muscle spindle")

    def muscle_spindle_control(self):
        pass

```

```python
# Importing all the required libraries
import numpy as np
import matplotlib.pyplot as plt

# parameters
frequency = 5  # Hz
sampling_interval = 0.01  # Time between every sample
t = np.arange(0, 1, sampling_interval)  # Time vector
signal = np.sin(2 * np.pi * frequency * t)  # Original sinusoidal signal

def downsample_signal(signal, downsample_rate=5, preserve_extrema=True):
    """
    Downsamples a signal while optionally preserving its maxima and minima.
    
    Parameters:
        signal (np.array): Original signal.
        downsample_rate (int): Downsample rate (default: every 5th point).
        preserve_extrema (bool): Whether to preserve maxima and minima (default: True).
    
    Returns:
        downsampled_signal (np.array): Downsampled signal.
        key_indices (list): Indices of the downsampled points.
    """
    # Identify max and min if required
    max_indices, min_indices = [], []
    if preserve_extrema:
        max_indices = [i for i in range(1, len(signal) - 1) if signal[i - 1] < signal[i] > signal[i + 1]]
        min_indices = [i for i in range(1, len(signal) - 1) if signal[i - 1] > signal[i] < signal[i + 1]]
    
    # Select downsampled indices
    sampled_indices = list(range(0, len(signal), downsample_rate))  # Downsample every 'downsample_rate' points
    key_indices = sorted(set(sampled_indices + max_indices + min_indices))  # Combine and remove duplicates
    
    # Create the downsampled signal
    downsampled_signal = signal[key_indices]
    
    return downsampled_signal, key_indices

# Downsample the signal
downsampled_signal, key_indices = downsample_signal(signal, downsample_rate=10, preserve_extrema=True)

# Plot the original and downsampled signals
plt.figure(figsize=(10, 6))
plt.plot(t, signal, label="Original Signal", color="blue", linewidth=2)
plt.scatter(t[key_indices], downsampled_signal, color="red", label="Downsampled Signal (Key Points)")
plt.title("Original vs Downsampled Signal")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Importing all the required libraries 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Linking the file path to the notebook and locating the required file
file_path = r"C:\Users\97254\.ssh\laptop_price - dataset.csv"
data = pd.read_csv(file_path)

class Laptop:
    # The constructor (__init__) initializes the laptop object with all necessary attributes
    def __init__(self, company, product, type_name, inches, screen_resolution, 
                 cpu_company, cpu_type, cpu_frequency, ram, memory, gpu_company, 
                 gpu_type, op_sys, weight, price):
        # Initialize the laptop's attributes with values provided during instantiation
        self.company = company  # Brand
        self.product = product  # Model
        self.type_name = type_name  # Type or category of the laptop
        self.inches = inches  # Size of the laptop's screen in inches
        self.screen_resolution = screen_resolution  # Resolution of the screen
        self.cpu_company = cpu_company  # Company that manufactures the CPU 
        self.cpu_type = cpu_type  # Specific type of CPU
        self.cpu_frequency = cpu_frequency  # CPU frequency in GHz
        self.ram = ram  # Amount of RAM in the laptop (in GB)
        self.memory = memory  # Storage capacity of the laptop 
        self.gpu_company = gpu_company  # Company that manufactures the GPU 
        self.gpu_type = gpu_type  # Specific type of GPU 
        self.op_sys = op_sys  # Operating system running on the laptop
        self.weight = weight  
        self.price = price

class Company:
    # The constructor initializes the company object with a name and an empty list of laptops
    def __init__(self, name):
        self.name = name  # Name of the company
        self.laptops = []  # List to hold all laptops produced by this company

    # Method to add a laptop to the company's list of laptops
    def add_laptop(self, laptop):
        self.laptops.append(laptop)  # Adds a Laptop object to the list of laptops produced by the company

    # Method to calculate the average price of all laptops produced by the company
    def average_price(self):
        # List comprehension to extract the price of each laptop
        prices = [laptop.price for laptop in self.laptops]
        # Return the mean price of all laptops using NumPy's mean function
        return np.mean(prices)

def plot_price_distribution(data):
# Creating a graph of laptop prices
    plt.figure(figsize=(10, 6))
    plt.hist(data['Price (Euro)'], bins=20, edgecolor='black')
    plt.title("Distribution of Laptop Prices")
    plt.xlabel("Price (Euro)")
    plt.ylabel("Frequency")
    plt.show()

# Calculating the average price for each company
average_price_per_company = data.groupby('Company')['Price (Euro)'].mean()
print(average_price_per_company)

# Finding the company with the highest average price
max_average_price_company = average_price_per_company.idxmax()
max_average_price_value = average_price_per_company.max()

print(f"The company with the most expensive laptop on average is {max_average_price_company} with an average price of {max_average_price_value} Euro.")

```

```python
# Converting the price column to a numeric format and handling NaN values
data['Price (Euro)'] = pd.to_numeric(data['Price (Euro)'], errors='coerce')

# Cleaning rows with missing values
prices = data['Price (Euro)'].dropna()

# Creating a simple price graph
plt.plot(prices)
plt.title('Laptop Prices')
plt.xlabel('Laptop Index')
plt.ylabel('Price (Euro)')
plt.show()
```

```python
# Fixing the names of the operating systems
data['OpSys'] = data['OpSys'].replace({
    'Windows 10': 'Windows', 
    'Mac OS X': 'MacOS', 
    'Windows 7': 'Windows'
})

# Displaying all unique operating systems
print(data['OpSys'].unique())
```

```python
import matplotlib.pyplot as plt

# Creating graphs by operating system
def plot_os_price_distribution(data):
    os_types = data['OpSys'].unique()
    
    for os_type in os_types:
        os_data = data[data['OpSys'] == os_type]
        plt.figure(figsize=(6, 4)) 
        plt.hist(os_data['Price (Euro)'], bins=20, edgecolor='black')
        plt.title(f"Price Distribution for {os_type}")
        plt.xlabel("Price (Euro)")
        plt.ylabel("Frequency")
        plt.show()

# Calling the function to display the graphs
plot_os_price_distribution(data)

```

```python
# The relationship between RAM and price
def plot_ram_price_relationship(data):
    data['RAM (GB)'] = data['RAM (GB)'].astype(float)
    plt.figure(figsize=(6, 4))
    plt.scatter(data['RAM (GB)'], data['Price (Euro)'], alpha=0.5)
    plt.title("Relationship between RAM and Laptop Price")
    plt.xlabel("RAM (GB)")
    plt.ylabel("Price (Euro)")
    plt.show()

# Calling the function to display the graph
plot_ram_price_relationship(data)

# Creating a new column for storage type
data['Storage type'] = data['Memory'].apply(lambda x: x.split(' ')[1] if isinstance(x, str) else None)

# Displaying the data after adding the column
print(data[['Memory', 'Storage type']].head())
```

## Q2:
 What types of processors are in the laptops and how do they affect the price?
 Grouped analysis (groupby) of the price by processor type, and a bar plot showing the average price for each processor type.

```python
def plot_cpu_type_vs_price(data):
    # Calculating the average price by CPU type
    cpu_price = data.groupby('CPU_Type')['Price (Euro)'].mean().sort_values(ascending=False)
    
    # Creating a bar chart
    plt.figure(figsize=(12, 6))
    cpu_price.plot(kind='bar', color='skyblue', edgecolor='black')
    plt.title("Average Laptop Price by CPU Type")
    plt.xlabel("CPU Type")
    plt.ylabel("Average Price (Euro)")
    plt.xticks(rotation=90, size=8)
    plt.show()

# Calling the function to display the graph
plot_cpu_type_vs_price(data)
```
